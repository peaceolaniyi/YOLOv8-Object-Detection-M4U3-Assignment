```python
# Cell 3: Download the dataset
from roboflow import Roboflow
rf = Roboflow(api_key="ZADbRNdMG0HdYoZNkYj5")
project = rf.workspace("huiyao-hu-sj18e").project("construction-ppe-detection")
version = project.version(1)
dataset = version.download("yolov8")
```

```python
# Cell 4: Train the model
from ultralytics import YOLO

# Load the 'Small' model for better accuracy (0.85 mAP goal)
model = YOLO('yolov8s.pt')

# Start training for 30 epochs
results = model.train(
    data=f"{dataset.location}/data.yaml", # This automatically finds the path!
    epochs=30,
    imgsz=640,
    plots=True
)
```
